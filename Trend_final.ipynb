###### API 구현 함수
- 선택한 카테고리의 일별 클릭량의 MAX 값이 100으로 설정된 상대적 값이 결과로 나타남
- 세번째 index인 돼지고기(카테고리매핑[3]].values[0])를 고정시켜, 모든 카테고리의 상대값의 기준이 되도록 함 

```python
# 네이버 api로 카테고리별 클릭량 가져오는 코드

def get_api(start, end, name1, name2, index = 0, cate = 4):
    세부카테 = 카테고리매핑[cate]
    
    index1 = 카테고리[카테고리[세부카테[0]] == name1].index[0]
    index2 = 카테고리[카테고리[세부카테[0]] == name2].index[0]
    
    cat = ''
    
    temp = '{"name":"' + 카테고리.loc[index1, 세부카테].values[0] + '","param":["' + str(int(카테고리.loc[index1,세부카테].values[1])) + '"]}'
    cat += temp+','
    temp = '{"name":"' + 카테고리.loc[index2, 세부카테].values[0] + '","param":["' + str(int(카테고리.loc[index2,세부카테].values[1])) + '"]}'
    cat += temp+','
    temp = '{"name":"' + 카테고리.loc[index, 카테고리매핑[3]].values[0] + '","param":["' + str(int(카테고리.loc[index, 카테고리매핑[3]].values[1])) + '"]}'
    cat += temp
    
    body = "{\"startDate\":\"" + start + "\",\"endDate\":\"" + end + "\",\"timeUnit\":\"date\",\"category\":[" +cat + "],\"device\":\"pc\",\"ages\":[\"20\",\"30\"],\"gender\":\"f\"}";
    
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    request.add_header("Content-Type","application/json")

    response = urllib.request.urlopen(request, data=body.encode("utf-8"))
    rescode = response.getcode()
    if(rescode == 200):
        response_body = response.read()
        return response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)
```

```python
카테고리4 = list(카테고리['카테고리명4'])

N = len(카테고리4)
cate_df = pd.DataFrame({'period':pd.date_range("2017-08-01", "2020-11-13")})
cate_df['period'] = cate_df['period'].astype(str)


for i in tqdm(range(0,N-1,2)):
    name1, name2 = 카테고리4[i], 카테고리4[i+1]
    ans = get_api("2017-08-01", "2020-11-13", name1, name2, index = 0, cate = 4)
    data = json.loads(ans)

    temp = pd.DataFrame(data['results'][0]['data'])
    temp = temp.rename({'ratio':data['results'][0]['title']}, axis = 'columns')
    cate_df = pd.merge(cate_df, temp, left_on = 'period', right_on = 'period', how = 'outer')

    temp = pd.DataFrame(data['results'][1]['data'])
    temp = temp.rename({'ratio':data['results'][1]['title']}, axis = 'columns')
    cate_df = pd.merge(cate_df, temp, left_on = 'period', right_on = 'period', how = 'outer')

    temp = pd.DataFrame(data['results'][2]['data'])
    temp = temp.rename({'ratio':data['results'][2]['title']}, axis = 'columns')
    cate_df = pd.merge(cate_df, temp, left_on = 'period', right_on = 'period', how = 'outer')
    
name = 카테고리4[-1]

ans = get_api("2017-08-01", "2020-11-13", name, name, index = 0, cate = 4)
data = json.loads(ans)

temp = pd.DataFrame(data['results'][0]['data'])
temp = temp.rename({'ratio':data['results'][0]['title']}, axis = 'columns')
cate_df = pd.merge(cate_df, temp, left_on = 'period', right_on = 'period', how = 'outer')

temp = pd.DataFrame(data['results'][1]['data'])
temp = temp.rename({'ratio':data['results'][1]['title']}, axis = 'columns')
cate_df = pd.merge(cate_df, temp, left_on = 'period', right_on = 'period', how = 'outer')

cate_df.head(10)
```

### (3) 단위 통일
- 2017년 8월 1일의 돼지고기 카테고리의 값들을 기준으로 잡아 모든 날짜의 식품 카테고리 검색값 단위를 통일함

```python
cate_l = list(cate_df.columns)

for i in tqdm(range(1, len(cate_l)-2, 3)):
    print(i, end = '')
    
    cate_df[cate_l[i + 2]] = cate_df[cate_l[i + 2]][0]

    cate_df[cate_l[i]] = cate_df[cate_l[i]] / cate_df[cate_l[i + 2]]
    cate_df[cate_l[i + 1]] = cate_df[cate_l[i + 1]] / cate_df[cate_l[i + 2]] 
    cate_df.drop([cate_l[i + 2]], axis = 'columns', inplace = True)
    
cate_df[cate_l[-1]] = cate_df[cate_l[-1]][0]
cate_df[cate_l[-2]] = cate_df[cate_l[-2]] / cate_df[cate_l[-1]]
cate_df.drop([cate_l[-1]], axis = 'columns', inplace = True)
```

### (4) 카테고리 축소
* 2020년 검색량이 0인 날이 5일 이하인 식품 카테고리만 사용 

```python
tmp_cate_df = ((cate_df[cate_df['period'] >= '2020-0-01'] == 0).sum() <= 5).to_frame()
use_cate = list(tmp_cate_df[tmp_cate_df[0] == True].index)

cate_df = cate_df[use_cate]

len(cate_df.columns)
```

### (5) 이상치 처리
* 이상치 제거를 위해 3일씩 묶고 평균을 내 일일 데이터에 저장

```python
cate_df_3days = pd.DataFrame(columns = cate_df.columns)
cate = cate_df.columns[1:]

for column in tqdm(cate):
    for i in range(1, len(cate_df)-1):
        day = cate_df.loc[i, 'period']
        total = cate_df.loc[i - 1, column] + cate_df.loc[i, column] + cate_df.loc[i + 1, column]
        cate_df_3days.loc[i - 1, ['period', column]] = [day, total / 3]

cate_df_3days.head()
```

### (1)-1 시계열 분석 : ARIMA

```python
data = pd.read_csv("2. shopping_api_cate4_전처리.csv", index_col = 0)
data.head()
```

```python
# 계절성을 반영하기 위해 5일씩 묶음 5*73 = 365

test_df = data[['period','아이스크림']]
test_df = test_df.set_index('period')
test_df.index = pd.date_range(start = '2017-08-02', end = '2020-11-12')

test_5day_df = test_df.resample('5D').mean()
test_5day_df.head()
```

```python
# arima의 모수인 p,d,q의 product를 구함
# 가장 낮은 aic를 갖는 모수들의 조합을 구함

p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 73) for x in list(itertools.product(p, d, q))]

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(test_5day_df,
                                            order = param,
                                            seasonal_order = param_seasonal,
                                            enforce_stationarity = False,
                                            enforce_invertibility = False)
            results = mod.fit()
            print('ARIMA{}x{}73 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
# arima 모형

mod = sm.tsa.statespace.SARIMAX(test_5day_df,
                                order = (1, 1, 1),
                                seasonal_order = (1, 1, 0, 73),
                                enforce_stationarity = False,
                                enforce_invertibility = False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
# 12번의 step을 예측 즉, 5*12 = 60일 예측

results.forecast(steps = 12).plot()
```

```python
pred = results.get_prediction(start = pd.to_datetime('2020-06-07'), dynamic = False)
pred_ci = pred.conf_int() #추정된 계수의 신뢰구간 계산
ax = test_5day_df['2017-08-02':].plot(label = 'observed')
pred.predicted_mean.plot(ax = ax, label = 'One-step ahead Forecast', alpha =.7, figsize = (14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color = 'k', alpha =.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

### (1)-2 시계열 분석 : RNN

```python
data = pd.read_csv("2. shopping_api_cate4_전처리.csv", index_col = 0)
data.head()
```

```python
# minmax 스케일러이용
# 이전 50일의 데이터를 이용해 다음날을 예측하는 방식

cate = '기타과자'
during = 50

scaler = MinMaxScaler()
```

```python
def make_dataset(cate = cate, during = during):
    data_bread = data[['period',cate]]
    train_data = pd.DataFrame(data_bread.loc[:1187, cate])
    test_data = pd.DataFrame(data_bread.loc[1188:, cate])
    
    train_data_sc = scaler.fit_transform(train_data)
    test_data_sc = scaler.transform(test_data)
    
    train_sc_df = pd.DataFrame(train_data_sc, columns = ['Scaled'], index = train_data.index)
    test_sc_df = pd.DataFrame(test_data_sc, columns = ['Scaled'], index = test_data.index)
    
    total_df = pd.concat([train_sc_df, test_sc_df])
    for i in range(1, during + 1):
        total_df['Scaled_{}'.format(i)] = total_df['Scaled'].shift(i)
    
    train_sc_df, test_sc_df = total_df[:1188], total_df[1188:]
    
    x_train = train_sc_df.dropna().drop('Scaled', axis = 1)
    y_train = train_sc_df.dropna()[['Scaled']]

    x_test = test_sc_df.dropna().drop('Scaled', axis = 1)
    y_test = test_sc_df.dropna()[['Scaled']]
    
    x_train = x_train.values
    x_test = x_test.values

    y_train = y_train.values
    y_test = y_test.values
    
    x_train_t = x_train.reshape(x_train.shape[0], during, 1)
    x_test_t = x_test.reshape(x_test.shape[0], during, 1)
    
    return x_train_t, y_train, x_test_t, y_test

def make_model(parmeter = 64, input_shape = (during,1), epochs = 100):
    model = Sequential()
    model.add(LSTM(64, input_shape=input_shape))  
    model.add(Dense(1, activation='linear')) 
    
    model.compile(loss = 'mean_squared_error', optimizer = 'adam')
    early_stop = EarlyStopping(monitor = 'val_loss', patience = 10, verbose = 1)

    model.fit(x_train, y_train, epochs = epochs, validation_data = (x_val, y_val), callbacks = [early_stop],
              batch_size = 20, verbose = 1)

    return model

def predict_trend(date = 30):
    result = []
    temp = x_test.copy()
    before = temp[-1]
    for i in range(date):
        after = model.predict(before.reshape(1,during,1))[0]
        before = np.concatenate((after.reshape(1,1), before[:during-1]))
        result.append(after[0])
    return result
```

```python
rnn_data = pd.DataFrame()
after_= []

categ = cate_df.columns[1:].to_list()

for index, i in enumerate(tqdm(data.columns[1:])):
    x_train, y_train, x_test, y_test = make_dataset(cate = i, during = 50)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state = 2020)
    model = make_model()
    print()
    print(i, model.evaluate(x_test, y_test))
    after_.append(predict_trend(date = 60))    # 향후 60일을 예측
    rnn_data[categ[index]] = after_[index]
```

```python
def scraping_th_default(query):
    browser = webdriver.Chrome()
    browser.maximize_window()

    # 옵션 : 기본뷰, 관련도순, 6개월 
    head_url = 'https://search.naver.com/search.naver?where=view&query='
    tail_url = '&sm=tab_viw.all&nso=so%3Ar%2Cp%3A6m%2Ca%3Aall&mode=normal'
    browser.get(head_url + query + tail_url)
    
    time.sleep(0.5)
    
    # 페이지 스크롤
    prev_height = browser.execute_script('return document.body.scrollHeight') # 기본 페이지 높이 저장
    while True:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight)') # 스크롤을 가장 아래로 내림
        time.sleep(2)
        curr_height = browser.execute_script('return document.body.scrollHeight') # 현재 페이지 높이 저장
        if curr_height == prev_height:
            break
        prev_height = curr_height

    time.sleep(0.5)
    
    html = BeautifulSoup(browser.page_source,'html.parser')

    blog = html.find_all(attrs = {'class' : 'bx _svp_item'})

    blog_df = pd.DataFrame(columns = ['식품명', '제목','해시태그'])

    blog_title = []
    blog_hashtag = []

    for i in range(len(blog)):
        blog_title = blog[i].find('a', {'class' : 'api_txt_lines total_tit'}).get_text()
        try:
            blog_hashtag = blog[i].find('div', {'class' : 'total_tag_area'}).get_text()
            blog_hashtag = blog_hashtag.replace('#', '').split()
        except:
            blog_hashtag = float('nan')
        blog_df.loc[i] = query, blog_title, blog_hashtag 
    
    browser.close()
    
    return blog_df
```

### (2)-1 식품 카테고리별 연관검색어 크롤링 : W2V 기반
- W2V을 통해 추출한 식품이 기존 판매 식품인지, 신규 식품인지 판단하기 위함
- 카테고리별 연관어를 네이버 쇼핑에 검색하여 검색되는 수를 크롤링함

```python
view_df = pd.read_csv('4. scraping_df_total.csv', index_col = 0)

view_df = view_df.fillna('[]')
view_df['해시태그'] = view_df['해시태그'].apply(literal_eval)
view_df.head()
```

```python
def search_count(query):
    site = 'https://search.shopping.naver.com/search/all?query=' + query
    req = requests.get(site)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    try:
        count = soup.find_all('span', attrs = {'class':'subFilter_num__2x0jq'})[0].text
        count = int(count.replace(',', ''))
    except:
        count = 0

    return count
```

```python
# 워드투벡 사용을 위해 리스트 안에 리스트 형태로, 제목과 해시태그는 명사만 추출
w2v_nouns = []
foods = ['가리비', '고구마', '두유', '만두', '알밤', '자몽', '케이크', '콜라겐', '쿠키', '홍차']

okt = Okt()

for food in foods:
    food_df = view_df[view_df['식품명'] == food]
    for i in food_df.index:
        w2v_nouns += [okt.nouns(food_df.loc[i,'제목'])]
        if not food_df.loc[i,'해시태그']:   # 해시태그가 없는경우
            continue
        w2v_nouns += okt.nouns(' '.join(food_df.loc[i,'해시태그']))
    
w2v_nouns[:10]
```

```python
max_len, min_len = 0, 0

for nouns in w2v_nouns:
    if not nouns: continue
    if len(nouns) > max_len:
        max_len = len(nouns)
    if len(nouns) < min_len:
        min_len = len(nouns)
        
max_len, min_len
```

```python
# 데이터가 적어서 CBOW 방식으로 최소 5개이상 단어만 사용 
trend_model = Word2Vec(sentences = w2v_nouns, min_count = 5, size = 100,  window = max_len, sg = 0, workers = 4)
```

```python
# 각 식품별로 유사도가 높은 순으로 500개씩 추출
w2v_df = pd.DataFrame(columns = ['W식품명', 'W연관어', 'W유사도'])

idx = 0

for food in tqdm(foods):
    related_food = trend_model.wv.most_similar(food, topn = 100000)
    for i in range(500):
        w2v_df.loc[idx] = [food] + list(related_food[i])
        idx += 1

w2v_df.head(10)
```

```python
# 크롤링

w2v_df['W검색수'] = 0
w2v_df['W검색어'] = ''

for i in tqdm(w2v_df.index):
    before = search_count(w2v_df.loc[i,'W연관어'] + w2v_df.loc[i,'W식품명'])  #앞에 연관어를 붙일 경우
    after = search_count(w2v_df.loc[i,'W식품명'] + w2v_df.loc[i,'W연관어'])  #뒤에 연관어를 붙일 경우

    if before >= after:
        w2v_df.loc[i, 'W검색수'] = before
        w2v_df.loc[i, 'W검색어'] = w2v_df.loc[i,'W연관어'] + w2v_df.loc[i,'W식품명']
    else:
        w2v_df.loc[i, 'W검색수'] = after
        w2v_df.loc[i, 'W검색어'] = w2v_df.loc[i,'W식품명'] + w2v_df.loc[i,'W연관어']
        
w2v_df = w2v_df[['W식품명', 'W연관어', 'W검색어', 'W유사도', 'W검색수']]
w2v_df.head(10)
```

### (2)-2 식품 카테고리별 연관검색어 크롤링 : Counter 기반
- Counter을 통해 추출한 식품이 기존 판매 식품인지, 신규 식품인지 판단하기 위함
- 카테고리별 연관어를 네이버 쇼핑에 검색하여 검색되는 수를 크롤링함

```python
view_df = pd.read_csv('4. scraping_df_total.csv', index_col = 0)

view_df = view_df.fillna('[]')
view_df['해시태그'] = view_df['해시태그'].apply(literal_eval)
view_df.head(10)
```

```python
# 제목과 해시태그에서 명사만 추출

def title_tag_counter(food):
    food_df = view_df[view_df['식품명'] == food].copy()

    tag_list = []
    for i in food_df.index:
        tag_list += food_df.loc[i,'해시태그']
    
    okt = Okt()
    nouns = []
    for i in food_df.index:
        nouns += okt.nouns(food_df.loc[i, '제목'])
    
    for i in tag_list:
        nouns += okt.nouns(i)
        
    total_count = Counter(nouns).most_common()
    return total_count
```

```python
def search_count(query):
    site = 'https://search.shopping.naver.com/search/all?query=' + query
    req = requests.get(site)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    try: 
        count = soup.find_all('span', attrs = {'class':'subFilter_num__2x0jq'})[0].text
        count = int(count.replace(',', ''))
    except:
        count = 0

    return count
```

```python
foods = ['가리비', '고구마', '두유', '만두', '알밤', '자몽', '케이크', '콜라겐', '쿠키', '홍차']
counter_df = pd.DataFrame(columns = ['C식품명', 'C연관어', 'C검색어', 'C태그수', 'C검색수'])

idx = 0

for food in tqdm(foods):
    related = title_tag_counter(food)
    for i in range(1, len(related)):
        if related[i][1] == 4:
            break
        
        before = search_count(related[i][0] + food)  # 앞에 연관어를 붙일 경우
        after = search_count(food + related[i][0])   # 뒤에 연관어를 붙일 경우
        
        if before >= after:
            counter_df.loc[idx] = food, related[i][0], (related[i][0] + food), related[i][1], before
        else:
            counter_df.loc[idx] = food, related[i][0], (food + related[i][0]), related[i][1], after
        
        idx += 1

counter_df.head(10)
```

- Counter를 통해 추출한 연관어의 W2V 기준 유사도를 df에 추가함 
  --> 이후에 유사도를 기준으로 최종 트렌드 식품을 선정하기 위함

```python
trend_model = Word2Vec(sentences = total_nouns, min_count = 5, size = 100,  window =  21, sg = 0, workers = 4)
tmp_df = pd.DataFrame(columns = ['C식품명', 'C연관어', 'C유사도'])
idx = 0

for food in tqdm(foods):
    related_food = trend_model.wv.most_similar(food, topn = 100000)
    for i in range(len(related_food)):
        tmp_df.loc[idx] = [food] + list(related_food[i])
        idx += 1

tmp_df.head(10)
```

```python
counter_df2 = pd.merge(counter_df, tmp_df, on = ['C식품명', 'C연관어'], how = 'outer')
counter_df2 = counter_df2[['C식품명', 'C연관어', 'C검색어', 'C태그수', 'C유사도', 'C검색수']]
counter_df2.head()
```

  ✔ 감성점수 = (긍정건수-부정건수) / 총건수 * 100

```python
def rank_crawl(keyword):
    # 키워드 분석    
    browser.implicitly_wait(10)
    
    browser.find_element_by_id('searchKeyword').clear()
    browser.find_element_by_id('searchKeyword').send_keys(keyword)
    browser.find_element_by_id('searchKeywordClick').send_keys(Keys.ENTER)

    time.sleep(10)

    # 검색조건 설정
    browser.find_element_by_xpath('//*[@id="wrap"]/div[1]/section/div/div/section[3]/div[2]/label[3]').click() #기간
    time.sleep(5)

    browser.find_element_by_xpath('//*[@id="keyword_section3"]/div[1]/div[1]/ul/li[2]').click()  #월별
    time.sleep(5)

    browser.find_element_by_xpath('//*[@id="keyword_section3"]/div[1]/div[2]/div/ul/li[3]/label').click()  #블로그
    time.sleep(3)
    browser.find_element_by_xpath('//*[@id="keyword_section3"]/div[1]/div[2]/div/ul/li[4]/label').click()  #인스타
    time.sleep(3)

    # 감성어 랭킹 크롤링
    rank_dec = browser.find_element_by_id('issueSentimentRank3').text.split('\n')

    browser.find_element_by_xpath('//*[@id="issueSentimentSlick"]/button[1]').click()
    rank_nov = browser.find_element_by_id('issueSentimentRank2').text.split('\n')

    browser.find_element_by_xpath('//*[@id="issueSentimentSlick"]/button[1]').click()
    rank_oct = browser.find_element_by_id('issueSentimentRank1').text.split('\n')

    browser.find_element_by_xpath('//*[@id="issueSentimentSlick"]/button[1]').click()
    rank_sep = browser.find_element_by_id('issueSentimentRank0').text.split('\n')

    rank_df = pd.DataFrame({'월':[], '순위':[], '분류':[], '키워드':[], '건수':[]})

    월 = []
    순위 = []
    분류 = []
    키워드 = []
    건수 = []

    rank_list = [rank_sep, rank_oct, rank_nov, rank_dec]
    rank_month = ['2020년 9월', '2020년 10월', '2020년 11월', '2020년 12월']

    for m in rank_list:
        for n in m:
            L = n.split(' ')
            월.append(rank_month[rank_list.index(m)])
            순위.append(L[0])
            분류.append(L[1])
            키워드.append(''.join(L[2:len(L)-1]))
            건수.append(L[len(L)-1])
        insert_data = pd.DataFrame({'월':월, '순위':순위, '분류':분류, '키워드':키워드, '건수':건수})

    rank_df = rank_df.append(insert_data)
    return rank_df
```

```python
top_df = pd.read_csv('6. top_df.csv', index_col = 0)

browser = webdriver.Chrome()
browser.maximize_window()
browser.implicitly_wait(10)
browser.get('https://some.co.kr/analysis/keyword')

browser.find_element_by_xpath('//*[@id="tutorial_wrap"]/div/ul/a').click()  #광고 제거

# 로그인
browser.find_element_by_xpath('//*[@id="leftsidebar"]/section/div/article[1]/div/a[1]').click()  
browser.find_element_by_id('username').send_keys('bangmir27@sogang.ac.kr')
browser.find_element_by_id('password').send_keys('insight1!')
browser.find_element_by_id('password').send_keys(Keys.RETURN)

# 크롤링
for i in tqdm(top_df.index):
    kw = top_df.loc[i, '연관어']
    df = rank_crawl(kw)
    
    df['건수'] = df['건수'].apply(lambda x: x.replace(',', ''))
    df['건수'] = df['건수'].apply(pd.to_numeric)
    
    총건수 = df['건수'].sum()
    긍정건수 = df[df['분류'] == '긍정']['건수'].sum()
    부정건수 = df[df['분류'] == '부정']['건수'].sum()
    # 점수 = (긍정-부정)/ 총건수
    점수 = (긍정건수-부정건수) / 총건수 * 100
    top_df.loc[i, '감성점수'] = 점수
```

### (5) 최종점수 도출
